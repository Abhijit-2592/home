```python
import pandas as pd

data_file = 'train_data.tsv'
df = pd.read_csv(data_file, error_bad_lines=False, delimiter='\t')
```

```python
df['phrase'].head()
```

```python
data = df.to_records().tolist()
```

```python
data[:10]
```

```python
import nltk

all_words = set()

for (_, __, phrase) in data:
    tokens = set(nltk.word_tokenize(phrase.lower()))
    all_words = all_words.union(tokens)

print all_words
```

```python
# now read stop words file

stop_words = set()

with open('SmartStoplist_modified.txt') as f:
    for line in f:
        # print line
        tokens = set(nltk.word_tokenize(line.lower()))
        stop_words = stop_words.union(tokens)

print stop_words
```

```python
# now find stop words available in target labels

good_words = all_words.intersection(stop_words)
```

```python
len(good_words)
```

```python
f2 = open('myfile','w')

with open('SmartStoplist_modified.txt') as f:
    for line in f:
        if line.rstrip() not in good_words:
            # write into bad words
            f2.write(line)
f2.close()
```
